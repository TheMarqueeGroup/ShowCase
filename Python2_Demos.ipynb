## Import Packages & Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# import plotly.io as pio
# pio.renderers.default='browser' #to launch graphs in browser when running in Spyder
# pio.templates.default = "plotly"
```

```python
#%% Import Data
allData = pd.read_excel('Output/allData.xlsx')
```

## Quick Pivots

```python
#%% Quick Pivots
pivotClient = allData.groupby(['Client Name'])[['Fees','$ Volume']].sum()
pivotClient['Fees'].plot(kind='bar')

pivotSectors = allData.groupby(['GICS Sector','GICS Sub Industry'])[['Fees','$ Volume']].sum()
```

```python
#%% Plotly Charts
fig = px.bar(allData, x='GICS Sector',  y='Fees', color='Client Name', hover_name='Ticker')
#fig.write_html(file="Output/sectors.html", auto_open=True)
fig.show()
```

```python
fig2 = px.bar(allData, x='Client Name',  y='Fees', color='GICS Sector')
fig2.write_html(file="Output/clients.html", auto_open=True)


figPie = px.sunburst(allData, values='Fees', path=['GICS Sector','GICS Sub Industry','Ticker'])
figPie.show()

figField = px.treemap(allData, values='Fees', path=['Client Name','GICS Sector','GICS Sub Industry','Ticker'])
figField.show()
```
